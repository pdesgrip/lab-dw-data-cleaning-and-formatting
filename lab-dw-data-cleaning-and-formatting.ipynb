## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they contain only valid data. 

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python

# 1. Cleans the Gender column
gender_new = {
    'Femal': 'F', 
    'Male': 'M', 
    'female': 'F', 
    'F': 'F',   # If it's already 'F', keep it as 'F'
    'M': 'M'    # If it's already 'M', keep it as 'M'
}
file1['gender'] = file1['gender'].map(gender_new).fillna(file1['gender'])

# 2. Cleans the State column
state_new = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington',
    'Arizona': 'Arizona',       # If it's already 'Arizona', keep it as 'Arizona'
    'California': 'California', # and so on...
    # Add more states if needed
}
file1['state'] = file1['state'].map(state_new).fillna(file1['state'])

# 3. Cleans the Education column
education_new = {
    'Bachelors': 'Bachelor',
    'Bachelor': 'Bachelor' # If it's already 'Bachelor', keep it as 'Bachelor'
}
file1['education'] = file1['education'].map(education_new).fillna(file1['education'])


# 5. Cleans Vehicle Class column
vehicle_class_new = {
    'Sports Car': 'Luxury', 
    'Luxury SUV': 'Luxury', 
    'Luxury Car': 'Luxury',
    'Luxury': 'Luxury'  # If it's already 'Luxury', keep it as 'Luxury'
}
file1['vehicle_class'] = file1['vehicle_class'].map(vehicle_class_new).fillna(file1['vehicle_class'])

# Display the cleaned dataframe
display(file1.head())
```

```python
# Find the indices of the rows with null values in the 'gender' column
null_indices = file2[file2['gender'].isnull()].index

# Calculate the number of null values
num_nulls = len(null_indices)

# Split that number in half
half_num_nulls = num_nulls // 2

# Assign half of the null values to 'M' and the other half to 'F'
file2.loc[null_indices[:half_num_nulls], 'gender'] = 'M'
file2.loc[null_indices[half_num_nulls:], 'gender'] = 'F'
```
